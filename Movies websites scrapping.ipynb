```python
#Egybest function

def Egybest():
    
    #Opening website
    while True:
        vpn_act = input("VPN Activated? (y/n)>")
        if vpn_act == 'y':
            print('VPN Activated...')
            break
        elif vpn_act == 'n':
            print('VPN Not Activated, Please activate to procced')
        else:
            print('Input Invalid')
            
    driver.get('https://egybist.online')
    #driver.maximize_window()
    print('\nWebsite Opened...')
    
    #Define function searching for movies
    def movies():
        while True:
            while True:
                # Handling search inputs
                movie_name = input("Movie name> ")
                if len(movie_name)>=3:
                    break
                if len(movie_name)<3:
                    print('Enter at least 3 characters')
                #elif movie_name == 'cancel':
                #    print('Search process canceled')
                    break
            
            #Forming search link
            x = movie_name.split()
            name = "%20".join(x)
            driver.get("https://egybist.online/search?query=" + name)
        
            #Search results
            search_res = driver.find_elements(By.CLASS_NAME,'title')
            search_links = driver.find_elements(By.CLASS_NAME,'block')
            
            titles = [search_res[title].get_attribute('innerHTML') for title in range(len(search_res))]
            p_links = [search_links[p_link].get_attribute('href') for p_link in range(len(search_links))]
            categories = [search_links[cat].get_attribute('href').split('/')[3] for cat in range(len(search_links))]
            
            #Defining titles and links lists
            movies = []
            p_links_m = []
            for i in range(len(categories)):
                if categories[i]=='movie':
                    movies.append(titles[i])
                    p_links_m.append(p_links[i])
            #Compining titles and links in 2d arrays for movies
            movies_plinks = [[movies[x] , p_links_m[x]] for x in range(len(movies))]
            #print(movies_plinks)

            #Result count
            res_count = len(movies_plinks)

            #print count of search results
            print('Found ' + str(res_count) + ' results:')
            
            #If search results one or more
            if res_count >= 1:
                for i in range(res_count):
                    result = movies_plinks[i][0]
                    print(str(i+1)+': '+ result)
                #Choose desired movie
                while True:
                    choice = int(input('\nChoose a movie (Enter number)>'))-1
                    if choice in range(res_count):
                        break
                    else:
                        print('Invalid input') 
                break

            #If threre is no result
            else:
                print("No such a movie :( try another one !")
        
        #Getting movie page
        page_link = movies_plinks[choice][1]
        driver.get(page_link)
        
        #Choose quality
        d_options_src = driver.find_element(By.XPATH, '/html/body/main/div/div[2]/div/article[5]/section').get_attribute('outerHTML')
        d_options = bs(d_options_src,'html.parser').find_all("div", {"class": "tr flex-start"})
        d_options_count = len(d_options)
        while d_options_count>=1:
            print(f'\nThere are {d_options_count} qualities choose one:')
            p_d_links=[]
            for i in range(d_options_count):
                resolution=driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[5]/section/div[{i+2}]/div[2]').get_attribute("innerHTML")
                quality=driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[5]/section/div[{i+2}]/div[1]').get_attribute("innerHTML")
                
                #Check div count
                size=''
                if len(d_options[0].find_all('div')) == 4:
                    size=driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[5]/section/div[{i+2}]/div[3]').get_attribute("innerHTML")
                    p_d_link=driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[5]/section/div[{i+2}]/div[4]/a').get_attribute('href')
                elif len(d_options[0].find_all('div')) == 3:
                    p_d_link=driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[5]/section/div[{i+2}]/div[3]/a').get_attribute('href')
                else:
                    print('Check div for webpage version')
                
                p_d_links.append(p_d_link)
                print(str(i+1) + ': ' + resolution + ' '+ quality + ' ' +size)
                print('   Download page link: '+p_d_link)
            
            while True:
                choice = input('\nChoose number assigned to desired quality to download or type (cancel) > ')
                if int(choice)-1 in range(len(p_d_links)):
                    driver.get(p_d_links[int(choice)-1])
            
                    #Check iframe existence
                    if len(driver.find_elements(By.TAG_NAME,'iframe')) != 0:
                        iframe = driver.find_element(By.XPATH, '/html/body/div[1]/div/iframe')
                        driver.switch_to.frame(iframe)
                        d_link=driver.find_element(By.XPATH,'/html/body/div/a')
                    elif len(driver.find_elements(By.TAG_NAME,'iframe')) == 0:
                        d_link=driver.find_element(By.XPATH,'/html/body/div[1]/div/a')
                    else:
                        print('Check div for webpage version')
                    
                    #print download link
                    link_text=d_link.get_attribute('href')
                    print('\nDownload Link:'+link_text)
                    #d_link.click()
                    #print('File Downloading...')
                    break
                elif choice == 'cancel':
                    print('Download canceld')
                    break
                else:
                    print('Invalid input')
            
                    #confirm download
                    #while True:
                    #    confirm = input("Confirm Download (y/n):")
                    #    if confirm == 'y':
                    #        link.click()
                    #        print('File Downloading...')
                    #        break
                    #    elif confirm == 'n':
                    #        driver.quit()
                    #        break
                    #    else:
                    #        print('Invalid input')"""
            break
    
        if d_options_count==0:
            print('No Download link in this website')

#----------------------------------------------------------------------------------------------------------------------------------------
    
    #Define function searching for series
    def series():
        #print("Feature under development\n")
        while True:
            while True:
                # Handling search inputs
                serie_name = input("serie name > ")
                if len(serie_name)>=3:
                    break
                if len(serie_name)<3:
                    print('Enter at least 3 characters')
                #elif serie_name == 'cancel':
                #    print('Search process canceled')
                    break
            
            #Forming search link
            x = serie_name.split()
            name = "%20".join(x)
            driver.get("https://egybist.online/search?query=" + name)
        
            #Search results
            search_res = driver.find_elements(By.CLASS_NAME,'title')
            search_links = driver.find_elements(By.CLASS_NAME,'block')
            
            titles = [search_res[title].get_attribute('innerHTML') for title in range(len(search_res))]
            p_links = [search_links[p_link].get_attribute('href') for p_link in range(len(search_links))]
            categories = [search_links[cat].get_attribute('href').split('/')[3] for cat in range(len(search_links))]
            
            #Defining titles and links lists
            series = []
            p_links_s = []
            for i in range(len(categories)):
                if categories[i]=='serie':
                    series.append(titles[i])
                    p_links_s.append(p_links[i])
            #Compining titles and links in 2d arrays for series
            series_plinks = [[series[x] , p_links_s[x]] for x in range(len(series))]
            #print(series_plinks)

            #Result count
            res_count = len(series_plinks)

            #print count of search results
            print('Found ' + str(res_count) + ' results:')
            
            #If search results one or more
            if res_count >= 1:
                for i in range(res_count):
                    result = series_plinks[i][0]
                    print(str(i+1)+': '+ result)
                #Choose desired movie
                while True:
                    choice = int(input('\nChoose a serie (Enter number)>'))-1
                    if choice in range(res_count):
                        break
                    else:
                        print('Invalid input') 
                break

            #If threre is no result
            else:
                print("No such a serie :( try another one !")

        #Getting serie page
        driver.get(series_plinks[choice-1][1])
        
        #Getting episode page links
        episodes_sel = driver.find_element(By.XPATH, '/html/body/main/div/div[2]/div/article[4]/div/div/div').get_attribute('outerHTML')
        ep_plinks_src = bs(episodes_sel,'html.parser').find_all("a", {"class": "block"})
        
        ep_plinks=[]
        for i in range(len(ep_plinks_src)):
            ep_plinks.append(ep_plinks_src[i].attrs['href'])
        
        for p in range(len(ep_plinks)):
            #Getting serie page
            page_link = ep_plinks[p]
            driver.get(page_link)
            
            #Getting download links for different qualities
            d_options_src = driver.find_element(By.XPATH, '/html/body/main/div/div[2]/div/article[4]/section').get_attribute('outerHTML')
            d_options = bs(d_options_src,'html.parser').find_all("div", {"class": "tr flex-start"})
            d_options_count = len(d_options)
            while d_options_count>=1:
                if d_options_count==1:
                    print(f'\nEpisode {p+1}: there is {d_options_count} quality link:')
                else:
                    print(f'\nEpisode {p+1}: there are {d_options_count} quality links:')
                print('Printing download links for different qualities....\n')
                
                p_d_links=[]
                
                for i in range(d_options_count):
                    resolution=driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[4]/section/div[{i+2}]/div[2]').get_attribute("innerHTML")
                    quality=driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[4]/section/div[{i+2}]/div[1]').get_attribute("innerHTML")
                    
                    #Check div count
                    if len(d_options[0].find_all('div')) == 4:
                        size = driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[4]/section/div[{i+2}]/div[3]').get_attribute("innerHTML")
                        p_d_link = driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[4]/section/div[{i+2}]/div[4]/a').get_attribute('href')
                    elif len(d_options[0].find_all('div')) == 3:
                        p_d_link = driver.find_element(By.XPATH,f'/html/body/main/div/div[2]/div/article[4]/section/div[{i+2}]/div[3]/a').get_attribute('href')
                    else:
                        print('Check div for webpage version')
                    
                    p_d_links.append(p_d_link)
                
                for d in range(len(p_d_links)):
                    driver.get(p_d_links[d])
                    
                    #Check iframe existence
                    if len(driver.find_elements(By.TAG_NAME,'iframe')) != 0:
                        iframe = driver.find_element(By.XPATH, '/html/body/div[1]/div/iframe')
                        driver.switch_to.frame(iframe)
                        d_link=driver.find_element(By.XPATH,'/html/body/div/a')
                    elif len(driver.find_elements(By.TAG_NAME,'iframe')) == 0:
                        d_link=driver.find_element(By.XPATH,'/html/body/div[1]/div/a')
                    else:
                        print('Check div for webpage version')
                    
                    #print download link
                    link_text=d_link.get_attribute('href')
                    print(str(d+1)+': '+link_text)
                            
                break
            if d_options_count==0:
                print('No Download links for this episode')

    
#======================================================================================================================================================

    #Choosing between movies or series    
    while True:
        # Handling inputs
        choose_m_s = input("Movie or Serie >")
        if choose_m_s=='Movie':
            movies()
            repeat=input('Another search? (y/n)> ')
            if repeat == 'y':
                continue
            elif repeat == 'n':
                break
            else:
                print('Input Invalid')
        elif choose_m_s=='Serie':
            series()
            repeat=input('Another search? (y/n)> ')
            if repeat == 'y':
                continue
            elif repeat == 'n':
                break
            else:
                print('Input Invalid')
        else:
            print('Input Invalid') 
```
